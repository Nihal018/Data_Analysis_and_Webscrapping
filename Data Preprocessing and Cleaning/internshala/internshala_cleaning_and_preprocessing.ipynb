### SALARY


```python
df.salary.value_counts()
```

```python
def remove_ctc_and_year(s):
    split1 = s.split("CTC: ")
    if split1[1] == "Competitive salary":
        return "Competitive salary"
    
    stripped1 = split1[1].strip("₹ ")
    split2 = stripped1.split(" /")
    stripped2 = split2[0].strip('$')

    return stripped2


def convert_package_to_min_salaries(salary):
    if(salary == 'Competitive salary'):
        return 'Competitive salary'
    
    if(salary.find('-') != -1):
        sal_1, sal_2 = salary.split('-')
        integer_sal_1 = int(sal_1.replace(',',''))
        integer_sal_2 = int(sal_2.replace(',',''))
        min_sal = min(integer_sal_1, integer_sal_2)
        return min_sal
    else:
        return int(salary.replace(',',''))

def convert_package_to_max_salaries(salary):
    if(salary == 'Competitive salary'):
        return 'Competitive salary'
    if(salary.find('-') != -1):
        sal_1, sal_2 = salary.split('-')
        integer_sal_1 = int(sal_1.replace(',',''))
        integer_sal_2 = int(sal_2.replace(',',''))
        max_sal = max(integer_sal_1, integer_sal_2)
        return max_sal
    else:
        return int(salary.replace(',',''))

def convert_package_to_mean_salaries(salary):
    if(salary == 'Competitive salary'):
        return 'Competitive salary'
    if(salary.find('-') != -1):
        sal_1, sal_2 = salary.split('-')
        integer_sal_1 = int(sal_1.replace(',',''))
        integer_sal_2 = int(sal_2.replace(',',''))
        mean_sal = np.average([integer_sal_1, integer_sal_2])
        return mean_sal
    else:
        return int(salary.replace(',',''))
```

```python
df.salary = df.salary.apply(remove_ctc_and_year)
```

```python

df["min_salary"] = df.salary.apply(convert_package_to_min_salaries)
df["max_salary"] = df.salary.apply(convert_package_to_max_salaries)
df["avg_salary"] = df.salary.apply(convert_package_to_mean_salaries)

```

```python
def remove_applicants(s):
    test_str = str(s)
    re_extract_num = re.compile('\d+')
    output = re_extract_num.findall(test_str)
    if(len(output) != 1):
        return 0
    else:
        return output[0]

df.applicants = df.applicants.apply(remove_applicants)
```

```python
df.applicants = pd.to_numeric(df.applicants)
```
